# Creating new features from date and converting type to int 

```python
df['bathrooms']= df['bathrooms'].astype(int)
df['year']= df['date'].str[:4]
df['month']= df['date'].str[4:6]
df['year']= df['year'].astype(int)
df['month']= df['month'].astype(int)
df['nbRooms']= df['bathrooms']+ df['bedrooms']
def Age1(a, b):
    if (a == 0):#(df['yr_renovated'] == 0):
        return 0
        
    else: # elif (df['yr_renovated'] >0):
        return a-b #  df['age1']=df['yr_renovated']- df['yr_built']
    
def Age2(a, b, c):
    if (a == 0):
        return b-c
        #df['age2']=df['year']- df['yr_built']
    else: #  (df['yr_renovated'] >0):
        return b-a
       # df['age2']=df['year']- df['yr_renovated']

df['age1']=df.apply(lambda row: Age1(row['yr_renovated'], row['yr_built']), axis=1)
df['age2']=df.apply(lambda row: Age2(row['yr_renovated'], row['year'],row['yr_built']), axis=1)
df['satisfaction']= df['age1']+ df['age2']

df.drop(['id','date'],axis=1, inplace=True)
df.head()
```

Creating seasons

```python
seasons = {(1, 12, 2): 1, (3, 4, 5): 2, (6, 7, 8): 3, (9, 10, 11): 4}
def season(ser):
    for k in seasons.keys():
        if ser in k:
            return seasons[k]

df['seasons']= df.month.apply(season)
```

# Multiple-linear regression

```python
X=df[['age1','year','bedrooms','bathrooms', 'sqft_living','floors','waterfront', 'view','grade','yr_built','lat', 'long', 'sqft_lot15','condition' , 'zipcode'
 ,'sqft_living15' ]]
 # selected features 
Y= df['price']# select the target survived
# split  data into train and test test_size =0.3
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.30, random_state=123)
```

```python
def normalize (df):
    features= df.columns
    for column in features:
        mu = df[column].mean()
        sigma = df[column].std()
        df[column] = (df[column] - mu)/sigma
```

```python
normalize(X_train)
```

```python
normalize(X_test)
```

```python
model=LinearRegression()   #build linear regression model
model.fit(X_train,Y_train)  #fitting the training data
Y_pred=model.predict(X_test) #testing our models performance
scoreMultiple=  metrics.r2_score(Y_test,Y_pred)
print("R squared",scoreMultiple)
resq = model.score(X_train,Y_train)
print("Rsq",resq)
```

- score of multi-linear regression is much better than linear regression as mentioned previously from correlation matrix there is more than one feature correlated with the price they can't be ignored predecting the price using only one feature
